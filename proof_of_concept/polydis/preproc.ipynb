```python
import torch
import numpy as np
import dataset as dtst
from model import DisentangleVAE

import pretty_midi

import os
```

```python
SAMPLE_LEN = 4
PATH_IN = "samples"
PATH_OUT = "samples_out"
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = DisentangleVAE.init_model(device)
model_path = 'result/models/disvae-nozoth_epoch.pt'  
model.load_model(model_path, map_location=device)

print("Model loaded")
```

```python
chords = np.zeros((7,36))
chords[:,24] = 1
for i, j in enumerate([0,2,4,5,7,9,11]):
    chords[i,j] = 1
    if j in [0,5,7]:
        chm = [0,4,7]
    if j in [2,4,9]:
        chm = [0,3,7]
    if j == 11:
        chm = [0,3,6]
    for c in chm:
        chords[i,12+(j+c)%12] = 1
print(chords)
```

```python
ds = [i for i in range(7)]
progs = [[i] for i in range(7)]
for step in range(SAMPLE_LEN-1):
    print(step)
    new_progs = []
    for prog in progs:
        for d in ds:
            new_progs.append(prog+[d])
    progs = new_progs
progs
```

```python
def get_sample(song):
    chord_path = "{}/{}/chord.npy".format(PATH_IN,song)
    mixed_path = "{}/{}/mixed.npy".format(PATH_IN,song)
    chord = np.load(chord_path)
    mixed = np.load(mixed_path)
    mixed = mixed.reshape(1,mixed.shape[0],mixed.shape[1])
    root = np.zeros((16,12))
    bass = np.zeros((16,12))
    root[[i for i in range(16)],chord[:,0].astype(int)] = 1
    bass[[i for i in range(16)],chord[:,-1].astype(int)] = 1
    chord = np.concatenate((root,chord[:,1:-1],bass),axis=1)
    return chord, mixed

def prog_to_chdmat(chords, prog, repeat=2):
    out = np.zeros((len(prog)*repeat, chords.shape[1]))
    for i in range(len(prog)*repeat):
        out[i,:] = chords[prog[i//repeat],:]
    return out

def notes_to_list(notes):
    out = []
    for note in notes:
        out.append([note.start,note.end,note.pitch,note.velocity])
    return out
```

### Remember to use the batches

```python
for song in os.listdir(PATH_IN):
    out = {}
    chord, pr_mat = get_sample(song)
    
    pr_mat = torch.from_numpy(pr_mat).float().to(device)
    chord = torch.from_numpy(chord).float().unsqueeze(0).to(device)
    
    for step in range(2):
        out_i = {}
        
        pr_mat_i = pr_mat[:,32*step:32*step+32,:]
        chord_i = chord[:,8*step:8*step+8,:]
        
        for i, prog in enumerate(progs):
            if i%100 == 0:
                print(i)
            
            chd_in = torch.from_numpy(prog_to_chdmat(chords, prog)).float().unsqueeze(0).to(device)
            
            ptree_out_i = model.swap(pr_mat_i, pr_mat_i, chord_i, chd_in, fix_rhy=True, fix_chd=False)
            pr_out_i, notes_out_i = model.decoder.grid_to_pr_and_notes(ptree_out_i.squeeze(0))


            out_i[str(prog)] = notes_to_list(notes_out_i)
        out[step] = out_i
    np.save("{}/{}.npy".format(PATH_OUT,song), out)
```
